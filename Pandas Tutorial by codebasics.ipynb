Concatenate Dataframe

```
bd_weather = pd.DataFrame({
    'city':['Dhaka', 'Comilla', 'Brahmanbaria'],
    'Temperature': [20, 30, 40],
    'Humidity': [70,80,55]

})
bd_weather
```

```
in_weather = pd.DataFrame({
    'city':['Dhaka', 'Kolkata', 'Agortala'],
    'Temperature': [22, 25, 35],
    'Humidity': [45,50,60]

})
in_weather
```

```
df = pd.concat([bd_weather, in_weather], keys=['bd', 'india'])
df
```

```
df.loc['bd']
```
