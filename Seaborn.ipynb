```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
df.dtypes
```

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

```python
## Jointplot
#Bivariate analysis

sns.jointplot(x='tip',y='total_bill',data=df, kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df, kind='reg')
```

```python
# pairplot

sns.pairplot(df)
```

```python
sns.pairplot(df, hue='sex')
```

```python
df['smoker'].value_counts()
```

```python
sns.pairplot(df, hue='smoker')
```

```python
## Dist plot - help you to create histogram
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False, bins=10)
```

```python
## Categorical plot
# boxplot
# violinplot
# countplot
# jontplot
# pairplot
```

```python
#count plot

sns.countplot(df['sex'])
```

```python
sns.countplot(df['day'])
```

```python
# barplot

sns.barplot(y='total_bill', x='sex',data=df)
```

```python
# Box plot

sns.boxplot('smoker','total_bill',data=df)
```

```python
sns.boxplot('sex','total_bill',data=df)
```

```python
sns.boxplot('day','total_bill',data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,orient='v')
```

```python
sns.boxplot(x='total_bill',y='day',hue='smoker', data=df)
```

```python
# viloin plot helps us to see both the distribution of data in terms of kernel
# density estimation and the box plot

sns.violinplot(x='total_bill',y='day',data=df,palette='rainbow')
```
