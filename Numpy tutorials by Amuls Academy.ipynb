Solve Linear Equation in Numpy ( Numerical Python)

3x + y = 9

x + 2y = 8

x=2, y=3

```
import numpy as np
a = np.array([[3,1],[1,2]])

b = np.array([9,8])
b
```

```
a
```

```
x = np.linalg.solve(a,b)
x
```
